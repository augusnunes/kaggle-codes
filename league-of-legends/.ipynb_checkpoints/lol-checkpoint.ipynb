```python
# Cor vermelha
red_c = "#db1d23"
# Cor azul
blue_c = "#1d5cdb"
```

```python
columns_blue = list(df.filter(regex="blue_*").columns)
columns_blue.append('match_duration')

df_blue = df[df.team_win == 'blue'].loc[:, columns_blue]
```

```python
columns_red = list(df.filter(regex="red_*").columns)
columns_red.append('match_duration')

df_red = df[df.team_win == 'red'].loc[:, columns_red]
```

```python
sns.displot(data=df,x='match_duration', hue='team_win', kind='kde', palette = [blue_c, red_c])
```
